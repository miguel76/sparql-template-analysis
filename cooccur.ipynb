```python
log_filename = 'data/execsByTemplateAndHostAndTime/bench-swdf-lsq2.csv'
logs = pd.read_csv(log_filename, sep=',', parse_dates=['execTime'])
```

```python
logs_by_cluster = load_log_file(log_filename, template_clusters, template_ids)
```

```python
for cluster_id, cluster_logs in enumerate(logs_by_cluster):
    if (cluster_id != 4):
        print(f'Working on cluster {cluster_id}...')
        event_log = pm4py.format_dataframe(cluster_logs, case_id='caseId', activity_key='templateId', timestamp_key='execTime')
        process_model = pm4py.discover_bpmn_inductive(event_log)
        pm4py.view_bpmn(process_model)
```
